```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs
```

```python
result = pd.read_csv('XSMB.csv')
result.head(5)
```

```python
result['odd_and_even'] = result['KQ'].apply(lambda x: 0 if x % 2 == 0 else 1)
result
```

```python
print(result)
```

```python
result['less than 1000'] = result['KQ'].apply(lambda x: 0 if x > 1000 else 1)
result
```

```python
result['odd_and_even'].value_counts()
```

```python
result['XX result'] = result['KQ'].apply(lambda x: x % 100)
result_new = result['XX result'].value_counts()
result_new

```

```python
xsmb_odd_even = result.groupby('odd_and_even').count().reset_index()
xsmb_odd_even 
```

```python
xx_result = result.groupby('XX result').count().reset_index()
sbs.barplot(data = xx_result, x = 'XX result', y = 'KQ')
```
